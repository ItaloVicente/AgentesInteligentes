```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
burro = pd.read_csv('robo_burro.csv')
inteligente = pd.read_csv('robo_inteligente.csv')
```

```python
burro, inteligente
```

```python
robo_burro = burro[burro['Iteração'] != 'Média']
robo_inteligente = inteligente[inteligente['Iteração'] != 'Média']

for col in ['Acertos', 'Erros', 'Movimentos Realizados']:
    robo_burro[col] = pd.to_numeric(robo_burro[col], errors='coerce')
    robo_inteligente[col] = pd.to_numeric(robo_inteligente[col], errors='coerce')

plt.figure(figsize=(14, 6))


plt.subplot(1, 3, 2)
plt.plot(robo_burro['Iteração'], robo_burro['Erros'], label='Robo Burro')
plt.plot(robo_inteligente['Iteração'], robo_inteligente['Erros'], label='Robo Inteligente')
plt.title('Comparação de Erros')
plt.xlabel('Iteração')
plt.ylabel('Erros')
plt.legend()
plt.xticks(ticks=range(0, len(robo_burro['Iteração']), 2)) 

plt.subplot(1, 3, 3)
plt.plot(robo_burro['Iteração'], robo_burro['Movimentos Realizados'], label='Robo Burro')
plt.plot(robo_inteligente['Iteração'], robo_inteligente['Movimentos Realizados'], label='Robo Inteligente')
plt.title('Comparação de Movimentos Realizados')
plt.xlabel('Iteração')
plt.ylabel('Movimentos Realizados')
plt.legend()
plt.xticks(ticks=range(0, len(robo_burro['Iteração']), 2))  # Mostrar apenas algumas marcas no eixo X

plt.tight_layout()
plt.show()
```
