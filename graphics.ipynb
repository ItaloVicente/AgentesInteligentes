```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
burro = pd.read_csv('results/aspirador_burro4x4.csv')
inteligente = pd.read_csv('results/aspirador_inteligente4x4.csv')

aspirador_burro = burro[burro['Iterao'] != 'Mdia']
aspirador_inteligente = inteligente[inteligente['Iterao'] != 'Mdia']

for col in ['Acertos', 'Erros', 'Movimentos Realizados']:
    aspirador_burro[col] = pd.to_numeric(aspirador_burro[col], errors='coerce')
    aspirador_inteligente[col] = pd.to_numeric(aspirador_inteligente[col], errors='coerce')

plt.figure(figsize=(14, 6))


plt.subplot(1, 3, 2)
plt.plot(aspirador_burro['Iterao'], aspirador_burro['Erros'], label='aspirador Burro')
plt.plot(aspirador_inteligente['Iterao'], aspirador_inteligente['Erros'], label='aspirador Inteligente')
plt.title('Comparao de Erros')
plt.xlabel('Iterao')
plt.ylabel('Erros')
plt.legend()
plt.xticks(ticks=range(0, len(aspirador_burro['Iterao']), 2)) 

plt.subplot(1, 3, 3)
plt.plot(aspirador_burro['Iterao'], aspirador_burro['Movimentos Realizados'], label='aspirador Burro')
plt.plot(aspirador_inteligente['Iterao'], aspirador_inteligente['Movimentos Realizados'], label='aspirador Inteligente')
plt.title('Comparao de Movimentos Realizados')
plt.xlabel('Iterao')
plt.ylabel('Movimentos Realizados')
plt.legend()
plt.xticks(ticks=range(0, len(aspirador_burro['Iterao']), 2))  # Mostrar apenas algumas marcas no eixo X

plt.tight_layout()
plt.show()
```
